```python
%matplotlib notebook
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow import keras
```

```python
mu, sigma = 2, 0.9
x1 = np.random.normal(mu, sigma, 100)
#x1 = np.arange(1,100)
x2 = np.random.normal(loc=x1, scale=2)
y = np.random.normal(loc=x1+x2, scale=1)
x = np.concatenate([x1.reshape(-1, 1), x2.reshape(-1, 1)], 1)
```

```python
model = tf.keras.Sequential([keras.layers.Dense(units=1, input_shape=[2])])
model.compile(optimizer='sgd', loss='mean_squared_error')
model.fit(x, y, epochs=500)
```
